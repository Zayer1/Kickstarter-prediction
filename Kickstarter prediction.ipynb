```
df['usd_pledged'] = pd.to_numeric(df['usd_pledged'], errors='coerce')
df['usd_pledged'].fillna(df['usd_pledged'].median(), inplace=True)
```

```
df = df.dropna(subset = ['name'])
```

```
# Correcting the syntax error in the code

df = df.dropna(subset=['category']).reset_index(drop=True)
```

```
df.isnull().sum()
```

```
df.dtypes
```

```
df['id'] = df['id'].astype('string')
df['name'] = df['name'].astype('string')
df['category'] = df['category'].astype('category')
df['main_category'] = df['main_category'].astype('category')
df['currency'] = df['currency'].astype('category')
df['deadline'] = pd.to_datetime(df['deadline'], errors = 'coerce')
df['goal'] = pd.to_numeric(df['goal'], errors = 'coerce')
df['launched'] = pd.to_datetime(df['launched'], errors = 'coerce')
df['pledged'] = pd.to_numeric(df['pledged'], errors = 'coerce')
df['state'] = df['state'].astype('category')
df['backers'] = pd.to_numeric(df['backers'], errors = 'coerce')
df['country'] = df['country'].astype('string')
```

```
df['duration_days'] = (df['deadline'] - df['launched']).dt.days
df['success'] = df['state'].apply(lambda x: 1 if x == 'successful' else 0)
df['launch_year'] = df['launched'].dt.year
df['launch_month'] = df['launched'].dt.month
df['launch_weekday'] = df['launched'].dt.day_name()
```

```
import seaborn as sns
import matplotlib.pyplot as plt
```

```
# Filter by state
filtered_df = df[df['state'].isin(['successful', 'failed'])]

# Remove extreme outliers
filtered_df = filtered_df[
    (filtered_df['goal'] > 0) & (filtered_df['goal'] < 100000) &
    (filtered_df['usd_pledged'] > 0) & (filtered_df['usd_pledged'] < 100000)
]

# Sample 5000 rows to avoid memory overload
sample_df = filtered_df.sample(n=5000, random_state=42)

# Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=sample_df, x='goal', y='usd_pledged', hue='success', alpha=0.5)
plt.xscale('log')
plt.yscale('log')
plt.title('Goal vs USD Pledged (Filtered + Sampled, Log Scaled)')
plt.xlabel('Goal (log scale)')
plt.ylabel('USD Pledged (log scale)')
plt.tight_layout()
plt.show()
```

```
df = df.dropna(subset = ['launched', 'deadline', 'goal', 'duration_days'])

```

```
df['backers'] = df['backers'].fillna(df['backers'].median())
df['pledged'] = df['pledged'].fillna(df['pledged'].median())
```

```
df['launch_year'] = df['launched'].dt.year
df['launch_month'] = df['launched'].dt.month
df['launch_weekday'] = df['launched'].dt.day_name()
```

```
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

# Features and target
features = ['goal', 'backers', 'duration_days']  # Add more like duration, category (encoded) later
target = 'success'

X = df[features]
y = df[target]

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train model
model = RandomForestClassifier()
model.fit(X_train, y_train)

# Evaluate
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```
